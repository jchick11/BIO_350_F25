```python
import matplotlib.pyplot as plt
```

```python
def N_log(N_12, k_400, time):
    # Setting up an logistic growth model
    N_values = [N_12]
    #Stores our initial N_value
    current_N = N_12
    #Sets current_N to the initial population for calculations
    carrying_capacity = k_400
    #Sets carrying_capacity to the carrying capacity for calculations
    for _ in range(time):
        #Sets a loop for the number of time steps
        death_rate = current_N * 0.4
        # This is our death rate equation that we will use to determine how many plants die each time step
        s = current_N * (1 - (current_N / carrying_capacity))
        #This is our growth rate equation that we will use to determine how many plants grow each time step
        current_N = current_N + (s - death_rate) * (1 - (current_N / carrying_capacity))
        #This is my logistic growth equation that I will use
        current_N = round(current_N, 0)
        #This is where we are rounding to O decimal places since its a whole plant
        if current_N > 400:
            current_N = 400
        # Sets the carrying capacity limit
        N_values.append(current_N)
        #This is how we build our list of N_values using our previous ones
    return N_values
    #This returns our list of N_values for future analysis
```

```python
result = N_log(12, 400, 25)
# This is where we call our function with our initial population of 12, carrying capacity of 400, and for 25 time steps
print(result)
#Gives us the results of our calculations
```

```python
years = list(range(26)) 
#Here we set up the range for our x axis for 25 years
plt.plot(years, result, marker='o')
#Here is where we define that the results we just printed are going to be plotted by each day and that the point will be marked with a circle
plt.xlabel('Years')
#Here we label the x axis
plt.ylabel('Plant Population')
#Here we label the y axis
plt.title('Plant Population Growth Over Time')
#Here we give the plot a title
plt.grid()
#Here we add a grid to the plot for easier reading
plt.show()
```
