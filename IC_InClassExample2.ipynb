Here we are importing and plotting population data

```python
# Import "Population_Data.csv" into a pandas DataFrame.
import pandas as pd
```

```python
#Import "Population_Data.csv" into a pandas DataFrame.
pop_data = pd.read_csv("Population_Data.csv")
pop_data.head()
#Average the population from days 17 to 60
avg_population = pop_data[(pop_data["Days"] >= 17) & (pop_data["Days"] <= 60)]["Population"].mean()
print(f"Average population from days 17 to 60: {avg_population}")
```

```python
#Plot the population data with "Days" on the x-axis and "Population" on the y-axis.
import matplotlib.pyplot as plt
plt.plot(pop_data["Days"], pop_data["Population"], marker='o', color='black', linestyle='-')
plt.axhline(y=276, color='green', linestyle='--', linewidth=2, label='Carrying Capacity')
plt.xlabel("Days")
plt.ylabel("Population")
plt.title("Population Growth Over Time")
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.grid(True)
plt.legend()
plt.show()
```
