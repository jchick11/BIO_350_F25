```python
# Import numpy
import numpy as np
```

```python
# Create an algorithm that calculates the population of three ages of a beetle population over 150 years
# The initial first year population is 500 and the initial second year population is 500
# On average one beetle produces 1 offspring in the second year and 1 individual produces 5 offspring in the third year
# The survival of first year beetles is 30% and the survival of second year beetles is 50%

# Create a definition to calculate and return the population of each age group and the total population for each year with deaths and births
def beetle_population(years, initial_first_year, initial_second_year):
    #Set explanation of parameters
    """
    Parameters:
    years (int): Number of years to simulate
    initial_first_year (int): Initial population of first year beetles
    initial_second_year (int): Initial population of second year beetles
    Returns:
    initial_first_year (list): Population of first year beetles for each year
    initial_second_year (list): Population of second year beetles for each year
    initial_third_year (list): Population of third year beetles for each year
    total_population (list): Total population for each year
    """
    # Create an assignment matrix
    A = np.array([[0, 1, 5], [0.3,0,0], [0,0.5,0]])
    n_t = np.array([initial_first_year, initial_second_year, 0.0], dtype=float)
    
    initial_population_vector = np.array([initial_first_year, initial_second_year, 0])
    #Create list to store population values
    initial_first_year = []
    initial_second_year = []
    initial_third_year = []
    total_population = []

    initial_first_year.append(n_t[0])
    initial_second_year.append(n_t[1])
    initial_third_year.append(n_t[2])
    total_population.append(sum(n_t))
    #Loop through the years to calculate population for each year
    n_t = initial_population_vector
    for year in range(1, years):
        n_t = np.dot(A, n_t)
        #Append current population to lists
        initial_first_year.append(n_t[0])
        initial_second_year.append(n_t[1])
        initial_third_year.append(n_t[2])
        total_population.append(sum(n_t))
        #Return population values
    return (initial_first_year, initial_second_year, initial_third_year, total_population)



```

```python
#Call the function with 50 years, 500 initial first year beetles, and 500 initial second year beetles
first_year, second_year, third_year, total_pop = beetle_population(150,500, 500)
```

```python
import matplotlib.pyplot as plt
```

```python
# plot populations for each age class and total
# ensure `years` matches the length of the population vectors
years = list(range(len(total_pop)))

plt.plot(years, first_year, marker='o', label='First-year', linestyle='-')
plt.plot(years, second_year, marker='s', label='Second-year', linestyle='--')
plt.plot(years, third_year, marker='^', label='Third-year', linestyle='-.')
plt.plot(years, total_pop, marker='x', label='Total', linestyle=':')
# make sure later code that references `result` doesn't error
plt.xlabel('Years')
plt.ylabel('Beetle Population')
plt.title('Beetle Population Growth Over Time')
plt.legend()
plt.grid()
plt.show()
```

```python
# Create an algorithm that calculates the population of three ages of a beetle population over 150 years
# The initial first year population is 500 and the initial second year population is 500
# On average one beetle produces 1 offspring in the second year and 1 individual produces 5 offspring in the third year
# The survival of first year beetles is 30% and the survival of second year beetles is 50%

# Create a definition to calculate and return the population of each age group and the total population for each year with deaths and births
def beetle_population(years, initial_first_year, initial_second_year):
    #Set explanation of parameters
    """
    Parameters:
    years (int): Number of years to simulate
    initial_first_year (int): Initial population of first year beetles
    initial_second_year (int): Initial population of second year beetles
    Returns:
    initial_first_year (list): Population of first year beetles for each year
    initial_second_year (list): Population of second year beetles for each year
    initial_third_year (list): Population of third year beetles for each year
    total_population (list): Total population for each year
    """
    # Create an assignment matrix
    trap_rate = 0.05
    A = np.array([[0, 1, 5], [0.3,0,0], [0,0.5,0]]) * (1 - trap_rate)
    n_t = np.array([initial_first_year, initial_second_year, 0.0], dtype=float)
    
    initial_population_vector = np.array([initial_first_year, initial_second_year, 0])
    #Create list to store population values
    initial_first_year = []
    initial_second_year = []
    initial_third_year = []
    total_population = []

    initial_first_year.append(n_t[0])
    initial_second_year.append(n_t[1])
    initial_third_year.append(n_t[2])
    total_population.append(sum(n_t))
    #Loop through the years to calculate population for each year
    n_t = initial_population_vector
    for year in range(1, years):
        n_t = np.dot(A, n_t)
        #Append current population to lists
        initial_first_year.append(n_t[0])
        initial_second_year.append(n_t[1])
        initial_third_year.append(n_t[2])
        total_population.append(sum(n_t))
        #Return population values
    return (initial_first_year, initial_second_year, initial_third_year, total_population)



```

```python
#Call the function with 50 years, 500 initial first year beetles, and 500 initial second year beetles
first_year, second_year, third_year, total_pop = beetle_population(150,500, 500)
```

```python
import matplotlib.pyplot as plt
```

```python
# plot populations for each age class and total
# ensure `years` matches the length of the population vectors
years = list(range(len(total_pop)))

plt.plot(years, first_year, marker='o', label='First-year', linestyle='-')
plt.plot(years, second_year, marker='s', label='Second-year', linestyle='--')
plt.plot(years, third_year, marker='^', label='Third-year', linestyle='-.')
plt.plot(years, total_pop, marker='x', label='Total', linestyle=':')
# make sure later code that references `result` doesn't error
plt.xlabel('Years')
plt.ylabel('Beetle Population')
plt.title('Beetle Population Growth Over Time')
plt.legend()
plt.grid()
plt.show()
```
