Creating an SIR model

```python
import matplotlib.pyplot as plt
```

```python
def sir_model(S_0, I_0, R_0, b, d, time):
    # Setting up an SIR model
    S_0 = 999 # Initial susceptible population
    I_0 = 1 # Initial infected population
    R_0 = 0 # Initial dead population
    b = 0.001  # Infection rate
    d = 0.5    # Death reate
    time = 30  # Time in days
    S = [S_0] # List to hold susceptible values over time
    I = [I_0] # List to hold infected values over time
    R = [R_0] # List to hold dead values over time

    for t in range(time):
        S_new = S[-1] - (b * S[-1] * I[-1]) # Susceptible at next time step
        I_new = I[-1] + (b * S[-1] * I[-1]) - (d * I[-1]) # Infected at next time step
        R_new = R[-1] + (d * I[-1]) # Dead at next time step
        S.append(S_new) # Append new values to list of susceptible
        I.append(I_new) # Append new values to list of infected
        R.append(R_new) # Append new values to list of dead
    return S, I, R # Return the lists of S, I, and R values over time
```

```python
resultS, resultI, resultR = sir_model(999, 1, 0, 0.001, 0.5, 30) # Call the function and store the results
print(resultS, resultI, resultR) # Print the results
```

```python
days = list(range(31))
#Here we set up the range for our x axis for 30 days
plt.plot(days, resultS, marker='o')
plt.plot(days, resultI, marker='o')
plt.plot(days, resultR, marker='o')
#Here is where we define that the results we just printed are going to be plotted by each day and that the point will be marked with a circle
plt.xlabel('Days')
#Here we label the x axis
plt.ylabel('Fox Susceptible, Infected, and Dead Population')
#Here we label the y axis
plt.title('Rabies Infection in Fox Population Over 30 Days')
#Here we give the plot a title
plt.legend(['Susceptible', 'Infected', 'Dead'])
#Here we give our plot a legend to differentiate between the three health states of the foxes
plt.grid()
#Here we add a grid to the plot for easier reading
plt.show()
```
