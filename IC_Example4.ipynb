```python
# Import numpy
import numpy as np
```

```python
# Create an algorithm that calculates the population of five ages of a vole population over 25 years
# The initial first year population is 50 and the initial second year population is 40, the initial third year population is 30, the 
# initial fourth year population is 20, and the initial fifth year population is 10
# On average one vole produces 3 offspring in the fourth year and 1 individual produces 2 offspring in the fifth year
# The survival of first year beetles is 80% and the survival of second year beetles is 85% the survival of third year beetles is 90% and the 
# survival of fourth year beetles is 79%
# Create a definition to calculate and return the population of each age group and the total population for each year with deaths and births
def vole_population(years, initial_first_year, initial_second_year, initial_third_year, initial_fourth_year, initial_fifth_year):
    #Set explanation of parameters
    """
    Parameters:
    years (int): Number of years to simulate
    initial_first_year (int): Initial population of first year beetles
    initial_second_year (int): Initial population of second year beetles
    Returns:
    initial_first_year (list): Population of first year beetles for each year
    initial_second_year (list): Population of second year beetles for each year
    initial_third_year (list): Population of third year beetles for each year
    total_population (list): Total population for each year
    """
    # Create an assignment matrix
    A = np.array([[0, 0, 0, 3, 2], [0.8, 0, 0, 0, 0], [0, 0.85, 0, 0, 0], [0, 0, 0.9, 0, 0], [0, 0, 0, 0.79, 0]], dtype=float)
    #Create initial population vector
    n_t = np.array([initial_first_year, initial_second_year, initial_third_year, initial_fourth_year, 0], dtype=float)
    
    initial_population_vector = np.array([initial_first_year, initial_second_year, initial_third_year, initial_fourth_year, 0])
    #Create list to store population values
    initial_first_year = []
    initial_second_year = []
    initial_third_year = []
    initial_fourth_year = []
    initial_fifth_year = []
    total_population = []

    initial_first_year.append(n_t[0])
    initial_second_year.append(n_t[1])
    initial_third_year.append(n_t[2])
    initial_fourth_year.append(n_t[3])
    initial_fifth_year.append(n_t[4])
    total_population.append(sum(n_t))
    #Loop through the years to calculate population for each year
    n_t = initial_population_vector
    for year in range(1, years):
        n_t = np.dot(A, n_t)
        #Append current population to lists
        initial_first_year.append(n_t[0])
        initial_second_year.append(n_t[1])
        initial_third_year.append(n_t[2])
        initial_fourth_year.append(n_t[3])
        initial_fifth_year.append(n_t[4])
        total_population.append(sum(n_t))
        #Return population values
    return (initial_first_year, initial_second_year, initial_third_year, initial_fourth_year, initial_fifth_year, total_population)
```

```python
#Call the function with 25 years, 50 initial first year beetles, and 40 initial second year beetles, 30 initial third year beetles,
#20 initial fourth year beetles, and 10 initial fifth year beetles
first_year, second_year, third_year, fourth_year, fifth_year, total_pop = vole_population(25,50,40,30,20,10)
```

```python
import matplotlib.pyplot as plt
```

```python
# plot populations for each age class and total
# ensure `years` matches the length of the population vectors
years = list(range(len(total_pop)))

plt.plot(years, first_year, marker='o', label='First-year', linestyle='-')
plt.plot(years, second_year, marker='s', label='Second-year', linestyle='--')
plt.plot(years, third_year, marker='^', label='Third-year', linestyle='-.')
plt.plot(years, fourth_year, marker='d', label='Fourth-year', linestyle='-')
plt.plot(years, fifth_year, marker='v', label='Fifth-year', linestyle='--')
plt.plot(years, total_pop, marker='x', label='Total', linestyle=':')
# make sure later code that references `result` doesn't error
plt.xlabel('Years')
plt.ylabel('Vole Population')
plt.title('Vole Population Growth Over Time')
plt.legend()
plt.grid()
plt.show()
```
